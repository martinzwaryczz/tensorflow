# Decorador @tf.function

Transforma las funciones de Python en operaciones de grafos, permitiendo mejorar el rendimiento y aprovechamiento de recursos.

Permiten añadir nuevas estructuras de control propias del framework:
- **tf.while_loop:** soporta **break** y **continue**: permite que múltiples iteraciones se ejecuten en paralelo si no tienen dependencias de datos.
- **tf.cond:** toma desiciones sobre el tensor en tiempo de ejecución.
- **tf.print:** imprime valores de tensores durante la ejecución del grafo

```python
@tf.function
def add(a,b):
  return a+b

x = tf.constant(5)
y = tf.constant(6)
z=add(x,y)
print(z)
```

## Arquitectura SEQUENTIAL

Documentación oficial: tensorflow.org/guide/keras/sequential_model

Es el modelo secuencial de capas compuestas por neuronas con una función de activación.

```python
# Importamos modulos

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Definición del modelo: generico, totalmente educativo no tiene sentido tenerlo así nomas

model_seq = Sequential([
    Dense(64, activation='relu', input_shape=(10,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid') # Capa de salida
])

# Compilación
model_seq.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model_seq.summary()
```

## Arquitectura con API

```python
# Importamos librerias

from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
# Definimos la entrada explícitamente, 100% generica para que ande
inputs = Input(shape=(10,))

# Conectamos las capas (el paréntesis al final indica de dónde vienen los datos)
x = Dense(64, activation='relu')(inputs)
x = Dense(32, activation='relu')(x)

# Definimos la salida
outputs = Dense(1, activation='sigmoid')(x)

# Creamos el modelo vinculando el inicio y el fin
model_func = Model(inputs=inputs, outputs=outputs)

# Compilación
model_func.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model_func.summary()
```

## Organizamos datos para crear y entrenar modelo

```python
# Datos de entrada para la arquitectura y sus funciones

import numpy as np # Si no lo importo acá no funciona

image_size=x_train.shape[1]
input_size=image_size*image_size
x_train=np.reshape(x_train,[-1,input_size])
x_train=x_train.astype('float32')/255
x_test=np.reshape(x_test,[-1,input_size])
x_test=x_test.astype('float32')/255
```

```python
batch_size=128 # Tamaño de lote
hidden_units=256 # Neuronas ocultas
dropout=0.45 # Fracción de las unidades de entrada que se desactivarán aleatoriamente durante el entrenamiento
```

## Creamos el modelo

```python
# Instanciamos el objeto Sequential()
model = Sequential()
```

```python
# Creamos la arquitectura

model.add(Dense(hidden_units, input_dim=input_size)) # Capa DENSE
model.add(Activation('relu')) # Capa con función de activación RELU
model.add(Dropout(dropout)) # Se van a apagar al azar el 48% de las neuronas
model.add(Dense(hidden_units)) # Capa DENSE "oculta" con 256 neuronas
model.add(Activation('relu')) # Capa con función de activación RELU
model.add(Dropout(dropout)) # Se van a apagar al azar el 48% de las neuronas
model.add(Dense(num_labels))
model.add(Activation('softmax')) # Capa de salida con función SOFTMAX
```

```python
# Resumen del modelo

model.summary()
```

```python
# Graficamos la arquitectura

plot_model(model,to_file='mlp.mnist.png',show_shapes=True)
```

## Compilamos el modelo

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy']) # Los fundamentos matemáticos son increíbles
```

## Entrenamos el modelo

```python
model.fit(x_train, y_train,epochs=20,batch_size=batch_size);
```

## Perdida y exactitud

```python
loss,acc = model.evaluate(x_test,y_test,batch_size=batch_size)
```

## Regularizador L2

 Técnica para prevenir el sobreajuste (overfitting) al penalizar los pesos grandes de una red neuronal

\section*{Regulación L2 en Keras}

La función de pérdida modificada con regularización L2 se expresa como:

La función de pérdida total con regularización L2 se define como:

$$L = L_{0} + \lambda \sum_{j=1}^{n} w_j^2$$

* **$L$**: Función de pérdida total final.
* **$L_{0}$**: Pérdida original (ej. Cross-Entropy).
* **$\lambda$**: Factor de regularización (el valor que definís en Keras).
* **$\sum w_j^2$**: Penalización por la suma de los pesos al cuadrado.
* **$w_j$**: Los pesos individuales de la capa.



```python
from tensorflow.keras.regularizers import l2
model.add(Dense(hidden_units,kernel_regularizer=l2(0.001),input_dim=input_size))
model.add(Dense(num_labels))
model.add(Activation('softmax'))
```

```python
# Estas capas fueron sumadas secuencialmente al modelo, la capa final sigue siendo SOFTMAX

model.summary()
```

### Compilamos otra vez

# Conceptos subyacentes de las arquitecturas


## Funciones de activación (Activation Functions)

Cada capa tiene una función de activación comprendida entre 0 y 1, hay muchas y se pueden ver en la documentación. NO SON EXACTAMENTE LAS MISMAS QUE PYTORCH.

## Optimizadores (Optim)

Algoritmos encargados de actualizar los pesos y sesgos de una red neuronal para minimizar la función de pérdida durante el entrenamiento. Su funcionamiento lleva consigo un gran e interesante fundamento matemático

## Función de perdida (Loss)

Cuantifica el error entre las predicciones del modelo y los valores reales. Conlleva también un gran e interesante fundamento matemático.

## Metricas de rendimiento (Model.fit(), Model.evaluate()y Model.predict())

Funciones que se utilizan para evaluar la calidad de las predicciones de tu modelo durante el entrenamiento y las pruebas.

# Mejorar el rendimiento de nuestros modelos

Para mejorar el rendimiento de nuestros modelos debemos:

- Seleccionar buenos optimizadores
- Usar funciones de activación correctas (evitar la linealidad)
- Evitar que sea dificultoso entrenar con SGD, este debe ser la brujula que nos lleve al error minimo: https://www.youtube.com/watch?v=k3qvYATh77I
- Normalizar los datos de entrada: https://www.baeldung.com/cs/normalizing-inputs-artificial-neural-network
- Usar reguladores L1 y L2.
- Crear capas con un dropout.
- Asignar un learning rate "razonable" (tamaño del paso que realiza el optimizador).
